```python
%%writefile app.py
from fastapi import FastAPI
import joblib
import pandas as pd

app=FastAPI()

pipeline=joblib.load("driver_drowsiness_pipeline.pkl")

@app.get("/")
def health_check():
  return {"status":"ok","message":"Drowsiness model is live"}

@app.post("/predict")
def predict_drowsiness(input_data:dict):
  df=pd.DataFrame([input_data])
  pred=pipeline.predict(df)[0]
  prob=pipeline.predict_proba(df)[0]

  return {
      "prediction":int(pred),
      "confidence":float(prob[1])
  }
```
